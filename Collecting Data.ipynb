## Fetch all genres from tMDB
We are getting all genres from tMDB because the kaggle dataset was collected using their API. 

```
api_key = "9d5ec76d7df373c1cca3e1aa89a82eff"
url = "https://api.themoviedb.org/3/genre/movie/list?api_key={}".format(
    api_key)
response = requests.get(url)
response.json()
```

```
df_genres = json_normalize(response.json()["genres"])
df_genres.to_csv("genres.csv")
df_genres.head()
```

```
url = "https://api.themoviedb.org/3/trending/movie/week?api_key={}".format(
    api_key
)
response = requests.get(url)
json_normalize(response.json()["results"])
```
